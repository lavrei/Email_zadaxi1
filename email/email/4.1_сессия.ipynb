```python

import pandas as pd
import textdistance
from tkinter import Tk, scrolledtext, Entry, Button

# Пример данных (замените его своими реальными данными)


df = data

def find_recipient_by_description(description):
    # Используем textdistance для нахождения наилучшего совпадения
    best_match = max(df["Краткое содежание"], key=lambda x: textdistance.jaccard(description.lower(), x.lower()))
    
    # Если совпадение достаточно высоко, возвращаем адресата
    if textdistance.jaccard(description.lower(), best_match.lower()) > 0.7:
        return df.loc[df["Краткое содежание"] == best_match, "Адресат"].values[0]
    else:
        return None

def issue_documents(period, recipient):
    # Ваш код для выдачи документов
    # (в данном примере просто возвращает список документов)
    return df[(df["дата регистрации"] == period) & (df["Адресат"] == recipient)]["Краткое содежание"].tolist() if recipient in df["Адресат"].values else None

def extract_params(text):
    # Пример извлечения параметров из текста команды
    params = {"description": None, "period": None, "recipient": None}
    
    if "определить адресата" in text:
        params["description"] = text.replace("определить адресата", "").strip()
    elif "выдать документы" in text:
        parts = text.split()
        params["period"] = parts[parts.index("дата") + 2] if "дата" in parts else None
        params["recipient"] = parts[parts.index("Адресат") + 1] if "Адресат" in parts else None
    
    return params

def execute_command(params):
    if params["description"]:
        recipient = find_recipient_by_description(params["description"])
        return f"Определен адресат: {recipient}" if recipient else "Адресат не найден"

    elif params["period"] and params["recipient"]:
        documents = issue_documents(params["period"], params["recipient"])
        if documents:
            return f"Выданы документы за период {params['period']} для адресата {params['recipient']}:\n{', '.join(documents)}"
        else:
            return f"Документы не найдены за период {params['period']} для адресата {params['recipient']}"

    elif params["description"] is None and params["period"] is None and params["recipient"] is None:
        return "Справка по командам:\n1. Определить адресата по описанию: 'определить адресата [описание]'\n2. Выдать документы за период по адресату: 'выдать документы дата регистрации [период] Адресат [адресат]'\n3. Справка: 'справка'"

    return "Неизвестная команда"

def process_user_input():
    user_input = input_entry.get()
    if user_input.lower() == "выход":
        root.destroy()
    else:
        params = extract_params(user_input)
        result = execute_command(params)
        output_text.insert("end", f"\nПользователь: {user_input}\nБот: {result}\n")
        output_text.yview("end")

# Создание графического интерфейса с использованием Tkinter
root = Tk()
root.title("Бот")

# Создание текстового поля для вывода результата
output_text = scrolledtext.ScrolledText(root, height=20, width=50)
output_text.pack()

# Создание поля для ввода команды
input_entry = Entry(root, width=50)
input_entry.pack()

# Создание кнопки "Отправить"
send_button = Button(root, text="Отправить", command=process_user_input)
send_button.pack()

root.mainloop()
```
