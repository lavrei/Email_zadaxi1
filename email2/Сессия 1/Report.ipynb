# 1.2 Работа с разреженным форматом данных


делаем словарь из test

```python
#first get all key value pairs
d = {}
for var in list(test.columns.values):
    d[var] = test[var].unique().tolist()

#then count
d_num = {}
for k,l in d.items():
    for v in l:
        d_num[(k+'_'+str(v))] = len(test[test[k]==v])


freqs = Counter(d_num.values()).most_common()
```
